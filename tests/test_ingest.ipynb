```python
import os

# Clases (carpetas)
CLASSES = ['Asma', 'Epoc', 'Neumonia', 'Normal']

#def ingest_data(base_path="data/raw"):
def ingest_data(base_path):
    data = []

    for label, clase in enumerate(CLASSES):
        class_path = os.path.join(base_path, clase)

        # Validación (muy importante)
        if not os.path.exists(class_path):
            print(f"⚠️ No existe la carpeta: {class_path}")
            continue

        for file in os.listdir(class_path):
            if file.endswith(".wav"):
                full_path = os.path.join(class_path, file)

                data.append({
                    "path": full_path,
                    "label": label,
                    "class_name": clase
                })

    print(f"✅ Total audios cargados: {len(data)}")
    return data
```

```python
# Ruta base donde están tus audios
#BASE_PATH = "..\\data\\raw" este funcioan en windows
BASE_PATH = "../data/raw"
```

```python
data = ingest_data(BASE_PATH)
```
